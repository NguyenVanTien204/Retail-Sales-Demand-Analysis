```python
from google.cloud import bigquery
from google.cloud import bigquery_storage
import pandas as pd
import os

TABLE_REF = "starry-lens-430815-s8.SummaryTable.Model"
OUTPUT_FILE = "Model_data.parquet"
```

```python
try:
    # Thiết lập client BigQuery tiêu chuẩn
    bqclient = bigquery.Client()
    # Thiết lập client BigQuery Storage (sử dụng để đọc tốc độ cao)
    bqstorageclient = bigquery_storage.BigQueryReadClient()
    print("Khởi tạo Client thành công!")
except Exception as e:
    print(f"Lỗi khởi tạo: {e}")
```

```python
try:
    dataframe = bqclient.list_rows(
        TABLE_REF
    ).to_dataframe(bqstorage_client=bqstorageclient)

    # Hiển thị 5 dòng đầu để kiểm tra nhanh
    display(dataframe.head())
except Exception as e:
    print(f"Lỗi khi đọc dữ liệu: {e}")
```

```python
try:
    print(f"Bắt đầu ghi dữ liệu ra file Parquet: {OUTPUT_FILE}...")
    dataframe.to_parquet(OUTPUT_FILE, index=False)

    # Kiểm tra kích thước file
    file_size_bytes = os.path.getsize(OUTPUT_FILE)
    file_size_mb = file_size_bytes / (1024 * 1024)

    print(f"Đã xuất dữ liệu thành công ra file {OUTPUT_FILE}.")
    print(f"Kích thước file: {file_size_mb:.2f} MB")
except Exception as e:
    print(f"Lỗi khi xuất file: {e}")
```

```python
try:
    df = dataframe

    unique_store_ids = df['store_id'].unique()

    output_dir = "state_id_files"
    os.makedirs(output_dir, exist_ok=True)

    print(f"Tìm thấy {len(unique_store_ids)} store_id duy nhất.")
    print(f"Thư mục lưu trữ: {output_dir}")
    print(f"Danh sách store_id: {unique_store_ids[:10]}...")
except Exception as e:
    print(f"Lỗi chuẩn bị: {e}")
```

```python
try:
    for store_id in unique_store_ids:
        store_df = df[df['store_id'] == store_id]

        store_file = os.path.join(output_dir, f"Model_data_store_{store_id}.parquet")

        store_df.to_parquet(store_file, index=False)
        print(f"Đã lưu: {store_file} ({len(store_df)} dòng)")

    print(f"Tất cả file đã được lưu trong thư mục '{output_dir}'.")
except Exception as e:
    print(f"Lỗi khi ghi file: {e}")
```
